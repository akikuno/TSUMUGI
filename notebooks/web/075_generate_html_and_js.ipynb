# index.html に表現型を代入


```python
mp_terms = {}
for path_mp_term in Path("data", "mp_term_name").glob("*.csv"):
    mp_term = path_mp_term.stem
    # print(path_mp_term.stem)
    if not Path("data", "network", f"{mp_term}.json").exists():
        continue
    mp_term_capitalized = mp_term.replace("_", " ")
    # print(mp_term_capitalized)
    mp_terms[mp_term_capitalized] = mp_term
```

# Template から HTML と JS を生成


```python
%%bash

########################################
# Clean up the old files
########################################

# cp web/network/template/template.html _template.html
# cp web/network/template/_template.js _template.js

rm web/network/html/*.html
rm web/network/js/*.js

# cp _template.html web/network/html/_template.html
# cp _template.js web/network/js/_template.js

########################################
# Extract the HTML and JS files
########################################

find data/mp_term_name -type f |
sed "s|data/mp_term_name/||" |
sed "s|.csv$||" |
while read mp_term; do
    mp_term_capitalized=$(echo $mp_term | awk '{ print toupper(substr($0, 1, 1)) substr($0, 2, length($0) - 1) }' | sed "s|_| |g")
    echo "Generating HTML and JS for $mp_term", capitalized as $mp_term_capitalized
    # HTML
    cat web/network/template/template.html |
    grep -v "REMOVE_THIS_LINE" |
    sed "s|XXX_capital_case|$mp_term_capitalized|g" |
    sed "s|XXX_snake_case|$mp_term|g" > web/network/html/$mp_term.html

    # Javascript
    cat web/network/template/template.js |
    grep -v "REMOVE_THIS_LINE" |    
    sed "s|XXX_snake_case|$mp_term|g" > web/network/js/$mp_term.js
done
```
