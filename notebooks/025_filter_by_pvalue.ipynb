```python
from pathlib import Path
from collections import defaultdict
from itertools import combinations
import csv
import pandas as pd
```

```python
path_data = Path("data", "impc", "statistical-results-ALL.csv")
```

```python
data = pd.read_csv(path_data) 
```

```python
print(len(data)) # 2062772
```

```python
# Filter by p_value < 0.0001
threshold = 0.0001
data = data[data["p_value"] < threshold]

# Filter by mp_term_id is not NaN
data = data.dropna(subset=["mp_term_id"])
data = data.dropna(subset=["mp_term_name"])

# Filter by effect_size is not NaN

data = data.dropna(subset=["effect_size"])
```

```python
data # 39356 rows 
```

```python
# NaNがあるか確認

data[data["marker_symbol"] == "Lepr"]["effect_size"].isnull().sum() # 0
```

```python
mp_terms = data["mp_term_name"].dropna().to_list() # True
mp_terms = list(set(mp_terms))
for mp_term in mp_terms:
    if "increased effector memory" in mp_term:
        print(mp_term)
    if "increased CD4" in mp_term:
        print(mp_term)
```

```python
data.to_csv("data/statistical_filtered.csv", index=False)
# data.to_csv("data/statistical_filtered.csv.gz", index=False, compression="gzip")
```
